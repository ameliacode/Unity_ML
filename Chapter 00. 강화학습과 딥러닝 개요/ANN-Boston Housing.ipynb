## Boston Housing Regression 
* Tensorflow 2.0 기준으로 작성
* Tensorboard 오류 해결 X → plot

```python
import tensorflow as tf
from sklearn import datasets
import numpy as np
import pandas as pd
import datetime
```

```python
algorithm = "ANN"
data_size = 13          #feature size
load_model = False

batch_size = 32         #num. data from train dataset
num_epoch = 500         #1 epochs per 32 batches

learning_rate = 1e-3

date_time = datetime.datetime.now().strftime("%Y%m%d-%H-%M-%S")

save_path = "./saved_models/"+date_time+"_"+algorithm
load_path = "./saved_models/"
```

```python
boston_housing = datasets.load_boston()
x = pd.DataFrame(boston_housing.data, columns = boston_housing.feature_names)
y = pd.Series(boston_housing.target)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(128, input_shape = (data_size,), activation = "relu"))
model.add(Dense(128, activation = "relu"))
model.add(Dense(128, activation = "relu"))
model.add(Dense(1))

model.compile(optimizer = tf.keras.optimizers.Adam(lr=learning_rate), loss="mse")
```

```python
# 모델 학습
history = model.fit(x_train, y_train, validation_data = (x_test,y_test), batch_size = batch_size, epochs = num_epoch)
```

```python
model.summary()
```

```python
model.evaluate(x_test,y_test)
```

```python
model.save(save_path)
print("Model is saved in {}".format(save_path + "/model/model"))
```

```python
import matplotlib.pyplot as plt

history_dict = history.history
print(history_dict.keys())

plt.figure(figsize=(15,8))

plt.xlabel("epoch")
plt.ylabel("loss")

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
